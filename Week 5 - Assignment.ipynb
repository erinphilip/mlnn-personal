# 1. Choose a REGRESSION dataset (reusing bikeshare is allowed), perform a test/train split, and build a regression model (just like in assignment 3), and calculate the
+ Training Error (MSE, MAE)
+ Testing Error (MSE, MAE)

```python
# linear regression using bikeshare

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 20, 10
import pandas as pd
import numpy as np
from sklearn import linear_model

day_hour_count = pd.read_csv("C:/Users/erinp/Desktop/mlnn/data/bikeshare_hour_count.csv")
day_hour_count
```

```python
from sklearn.model_selection import train_test_split
```

```python
# using saturday since no null values
saturday = day_hour_count[["hour", "saturday"]].copy()
```

```python
x_train, x_test, y_train, y_test = train_test_split(saturday["hour"].values.reshape(-1,1),saturday["saturday"], test_size=.20)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
model = linear_model.LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
# train MSE
from sklearn import metrics

metrics.mean_squared_error(y_train, model.predict(x_train))
```

```python
# train MAE
metrics.mean_absolute_error(y_train, model.predict(x_train))
```

```python
# test MSE
metrics.mean_squared_error(y_test, model.predict(x_test))
```

```python
# test MAE
metrics.mean_absolute_error(y_test, model.predict(x_test))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (20, 6)
plt.rcParams['font.size'] = 14
import pandas as pd
```

```python
heart = pd.read_csv("C:/Users/erinp/Desktop/mlnn/data/Heart.csv")
heart
```

```python
heart.columns
```

```python
non_numeric_columns = ['Unnamed: 0', 'ChestPain','Thal']
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion='entropy')
```

```python
x = heart.copy().drop(non_numeric_columns, axis=1)
x
```

```python
# drop NaN values in Ca
x = x.dropna()
x
```

```python
model.fit(x.drop(['AHD'], axis=1), x.AHD)
```

```python
list(zip(x.drop(['AHD'], axis=1).columns, model.feature_importances_))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x.drop(['AHD'], axis=1),x.AHD, test_size=.20)
```

```python
x.shape, x_train.shape, x_test.shape
```

```python
model.fit(x_train, y_train)
```

```python
test_predictions = model.predict(x_test)
```

```python
test_predictions
```

```python
y_test
```

```python
from sklearn.metrics import (accuracy_score, 
                             classification_report, 
                             confusion_matrix
                            )
```

```python
accuracy_score(y_test, test_predictions)
```

```python
confusion_matrix(y_test, test_predictions)
```
